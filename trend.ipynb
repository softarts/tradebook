# 短期趋势-移动平均指标
通过ma10，价格等揭示股票的趋势

1. ma10 指价格上穿ma10的bar数
2. slo10 指ma10的斜率
3. sth 指短期新高(20),lth指长期新高(200)

# scan criteria
1. ma10 出现信号;斜率10大于斜率20（即ma10上穿ma20）;斜率20大于斜率50
2. ma10 出现信号，sth短期新高信号少于8
3. ma10,20,50都出现信号
4. 现价大于sth短期新高的98%
5. 价格在98% - 102% ma200之间(即接近200天均线的股票)

# column
'symbol', 'px', 'macd', 'ma10%', 'ma20%', 'ma50%', 'ma10', 'ma20', 'ma50', 'ma200', 'volra20', 'vol20_sig', 'ma10_sig', 'ma20_sig', 'ma50_sig', 'slo10', 'slo20', 'slo50', 'sth', 'lth', 'stl', 'ltl', 'lth_sig', 'sth_sig', 'ltl_sig', 'stl_sig'


# sort
1. ma10_sig, 升序
2. ma50_sig, 升序

'sort': [{'ma10_sig': 'True'}, {'ma50_sig': 'True'}]}
