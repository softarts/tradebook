# SCTR
SCTR是用来表达一个股票在一组股票里的相对强度，sctr_rank 为100表明得分最高的那支股票 https://school.stockcharts.com/doku.php?id=technical_indicators:sctr

```python
from tps import common
from tps import tradebook as tb
import pandas as pd
import re
```

# load ohlc

```python
g_symbol_df = common.load_symbol_csv('./symbol_ib.txt')
ohlc_dct = tb.get_symbol_ohlc(g_symbol_df,fpath='./cache_example/')
```

```python
ohlc_dct
```

# load algorithm config

```python
cfg="""{
"algo":[
    {
      "name":"sctr",
      "param":{}
    }
  ]
}"""


algo_dct=tb.load_algo_pipeline(cfg)
```

# Processing algorithm 

```python
df=tb.run_algo(ohlc_dct,algo_dct)
```

```python
df
```

# 扫描规则
找出在逐步提升排名的股票

```python
cfg = """{  
  "criteria":[
    "sctr_rank>sctr_r5",
    "sctr_r5>sctr_r10"
  ],
  "columns":[
    "symbol","px",
    "stok","sctr","sctr_rank","sctr_r5","sctr_r10","sctr_r30"
  ],
  "sort":[
    {"sctr_rank":"False"}
  ]
}"""
#import json
#criteria=json.loads(cfg)
df_scan=tb.run_scan(df,cfg)
```
